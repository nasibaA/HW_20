```python
import nltk 
# books
nltk.corpus.gutenberg.fileids()
```

```python
# stopwords
sw = set(nltk.corpus.stopwords.words('english'))
# adding new meaningless words
sw.add('thee')
sw.add('one')
sw.add('u')
sw.add('thus')
sw.add('yet')
print(sw)
```

```python
# Paradise lost book loading as sentences in side list 
text_sentences = nltk.corpus.gutenberg.sents('milton-paradise.txt')
text_sentences
```

```python
#lower casing the all latter inside text and removing the stopwords
filter_words=[]
for sentence in text_sentences:
    filtered_list = [w.lower() for w in sentence if w.lower() not in sw]
# filtering alnumeric sentences and adding the filter_words list     
    new_filtered_list = [w for w in filtered_list if w.isalnum()]
    filter_words.append(new_filtered_list)
for e in filter_words:
    print(e)
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import regexp_tokenize, TweetTokenizer
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
from matplotlib import pyplot as plt

# Instantiate the WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

# taking filter_words and getting meaning words
lem=[]
for w in filter_words:
    lem.append([wi for wi in map(wordnet_lemmatizer.lemmatize, w)])
# counting the frequencies words    
print(lem)
frequencies = (word for sentence in lem for word in sentence)
frequencies_word = Counter(frequencies)
# taking most_common 20 words
for word,frequency in frequencies_word.most_common(20):
    print(word,frequency)
    
# making plot bar to make easy visualitation    
    plt.bar(word,frequency)
    plt.xticks(rotation=90)
```
